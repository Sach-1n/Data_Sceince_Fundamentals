##  Linear Regression

```python
import pandas as pd

df = pd.read_csv('rods.csv', sep = '\t')
print(df.head())
```

```python
import matplotlib.pyplot as plt



x = df['Rough_Weight_x'].to_list()
y = df['Finished_Weight_y'].to_list()

plt.figure(figsize = (10,7))
plt.scatter(x,y)
plt.title('Scatter Plot', size = 18)
plt.xlabel('Rough Weight',size = 10)
plt.ylabel('Finished Weight',size = 10)
plt.show()
```

```python
# Linear Regression
Sum = 0
for i in x:
    Sum += i;
x_mean = Sum/len(x);

Sum = 0
for i in y:
    Sum += i;
y_mean = Sum/len(y);

nume = 0
deno = 0
for i in range(len(x)):
    nume += (x[i]-x_mean)*(y[i]-y_mean)
    deno += pow((x[i] - x_mean),2)

# y = mx + c
m = nume/deno 
c = y_mean - m*x_mean

import numpy as np

xx = np.array([2.5,2.75])
plt.figure(figsize = (10,7))
plt.plot(xx, xx*m + c, color = 'red')

plt.scatter(x,y)
plt.title('Regression Line in Scatter Plot', size = 18)
plt.xlabel('Rough Weight',size = 10)
plt.ylabel('Finished Weight',size = 10)
plt.show()
```

```python
# Lets Assume the Line is in form : y = m*x + c
# Here the two Coefficients are m and c

print("\nSlope =", m)
print("\nIntercept Term =", c)
```

## Plotting the Residuals

```python
df = pd.read_csv('dataset.csv')
print(df.head())

x = df['Brain Weight'].to_list()
y = df['Head Size'].to_list()

# Linear Regression
Sum = 0
for i in x:
    Sum += i;
x_mean = Sum/len(x);

Sum = 0
for i in y:
    Sum += i;
y_mean = Sum/len(y);

nume = 0
deno = 0
for i in range(len(x)):
    nume += (x[i]-x_mean)*(y[i]-y_mean)
    deno += pow((x[i] - x_mean),2)

# y = mx + c
m = nume/deno 
c = y_mean - m*x_mean

res = y
for i in range(len(y)):
    res[i] -= m*x[i] + c

print()

plt.figure(figsize = (10,7))
plt.scatter(x,res)
plt.title('Residual Graph',size = 17)
plt.xlabel('Brain Weight',size = 13)
plt.ylabel('Residue',size = 13)
plt.axhline(y=0, color='r', linestyle='-')
plt.show()

```

## Polynomial Regression

```python
df= pd.read_csv('dataset.csv')  
  
x= df.iloc[:, 3:4].values  
y= df.iloc[:, 2].values  

# Fitting the Polynomial Regression
from sklearn.linear_model import LinearRegression  
from sklearn.preprocessing import PolynomialFeatures  
poly_regs= PolynomialFeatures(degree = 2)  
x_poly= poly_regs.fit_transform(x)  
lin_reg_2 =LinearRegression()  
lin_reg_2.fit(x_poly, y) 

# Plotting for Polynomial Regression  
plt.figure(figsize = (10,7))
plt.scatter(x,y)  
plt.plot(x, lin_reg_2.predict(poly_regs.fit_transform(x)), color="red")  
plt.title('Polynomial Regression',size = 17) 
plt.xlabel('Brain Weight',size = 13) 
plt.ylabel('Head Size',size = 13) 
plt.show()  
```
