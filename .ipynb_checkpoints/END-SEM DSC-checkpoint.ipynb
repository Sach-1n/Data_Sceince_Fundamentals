#### Q2 (a), (b), (c) Find q1, q2, q3

```python
import math as m

total = len(li)

li.sort()
print("LIST li =",li)

q1 = (total+1)/4
q2 = 2*q1
q3 = 3*q1

Q1 = (li[m.floor(q1)-1] + li[m.ceil(q1)-1])/2
Q2 = (li[m.floor(q2)-1] + li[m.ceil(q2)-1])/2
Q3 = (li[m.floor(q3)-1] + li[m.ceil(q3)-1])/2

print()
print()
print('First Quartile, Q1 =',Q1)
print()
print('Second Quartile, Q2 =',Q2)
print()
print('Third Quartile, Q3 =',Q3)
print()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(9, 10)) 
plt.boxplot(li)
plt.scatter(1, Q1, color = 'Red')
plt.annotate('Q1',(1, Q1), size = 15)
plt.scatter(1, Q2, color = 'Red')
plt.annotate('Q2',(1, Q2), size = 15)
plt.scatter(1, Q3, color = 'Red')
plt.annotate('Q3',(1, Q3), size = 15)
plt.xlabel('PLOT')
plt.ylabel('DATA')
plt.show()
```

#### 2 (g) What is mean?

Mean is value which we get by adding all the values of dataset and dividing it by the frequency.

```python
li_sum = 0
for i in li:
    li_sum = i + li_sum
    
mean = li_sum/total
print("MEAN", mean)
```

#### 2 (h) What is median?
Median is a value separating the higher half from the lower half of a data sample.

```python
print("MEDIAN", median)
```

#### 2 (i) What is mode?
The value with the highest number of frequency is mode.

```python
count = {}
for i in li:
    try :
        count[i] += 1
    except KeyError:
        count[i] = 1
maxx = 0
for i in count:
    if count[i] > maxx:
        maxx = count[i]
        j = i
        
        
        

print('\nMode = ',j)
    
```

#### 2 (j) Variance and Standard Deviation


```python
# Calculating the Variance
summ = 0
for i in li:
    val = i - mean
    summ += pow(val,2)
    
variance = summ / (total - 1)

print("Variance =",variance)


sd = m.sqrt(variance)
print("\nStandard Deviation =",sd)
```

#### Q2  (B) Create a program that computes the matrix multiplication of 2 matrices, using the recursion methods, use numpy library.


```python
from numpy import array
i = 0
j = 0
k = 0
def Matrix_Mul(row1, col1, A, row2, col2, B, C):  
        
    if (i >= row1):  
        return
          
    if (j < col2): 
        if (k < col1): 
            C[i][j] += A[i][k] * B[k][j] 
            k += 1
            Matrix_Mul(row1, col1, A,  
                              row2, col2,B, C) 
  
        k = 0
        j += 1
        Matrix_Mul(row1, col1, A,row2, col2, B, C) 
  
    j = 0
    i += 1
    Matrix_Mul(row1, col1, A,row2, col2, B, C)
    
A = array([[2,4],[4,2]])
B = array([[6,1],[8,5]])

(row1,col1) = A.shape
(row2,col2) = B.shape


C = array([[6,1],[8,5]])

Matrix_Mul(row1, col1, A, row2, col2, B, C)

print(C)
```
