## Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd

df2 = pd.read_csv("test_batting.csv", nrows = 300)
print(df2)
print()
matches = df2.Mat.tolist()
hundreds = df2.Hundred.tolist()

fig = plt.figure(figsize = (10, 7)) 
plt.scatter(matches,hundreds)
plt.xlabel("Matches Played", size = 13) 
plt.ylabel("Hundreds Scored" , size = 13) 
plt.title("Matches Played vs Hundreds Scored", size = 18) 
plt.show()


```

## Bar Plot

```python
df = pd.read_csv("most_runs.csv")
print(df)
print()
Batsman = df.Batsman.tolist()
Runs = df.Runs.tolist()

fig = plt.figure(figsize = (8, 5)) 
plt.bar(Batsman, Runs)
plt.xlabel("Batsman", size = 20) 
plt.ylabel("Runs" , size = 20) 
plt.title("Most Runs Scored IPL 2020", size = 20) 
plt.legend(["Runs Scored"])
plt.show()
```

## Histogram

```python
run_bin = [0, 1, 2, 4, 6]
count = [4, 0, 0, 1, 1, 1, 4, 0, 4, 1, 1, 6, 1, 1, 4 , 4, 0]

print()
fig = plt.figure(figsize = (9, 6)) 
plt.hist(count , bins = run_bin)
plt.xlabel("Runs", size = 20) 
plt.ylabel("Count" , size = 20) 
plt.title("Type of Runs Scored in First 3 Overs", size = 15) 
plt.legend(["Count"])
plt.show()
```

## Pie-Plot

```python
course = ['B.Tech','M.Tech CSE', 'M.Tech CSIS', 'PHD']
count = [240, 46, 42, 45]
print()
fig = plt.figure(figsize = (7, 7)) 
plt.pie(count, labels = course, autopct='%1.1f%%')
plt.title("Student Distribution in CSE Department", size = 15) 
plt.legend(["B.Tech","M.Tech CSE","M.Tech CSIS","PHD"], loc = 1)
plt.show()
```
